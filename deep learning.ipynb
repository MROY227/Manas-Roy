```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('/content/177317709_522893839096042_3703613067965651809_n.jpg')
```

```python
if image is None:
  print("Error:could not open or find the image.")
  exit()
```

```python
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

```python
sobel_y=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
```

```python
feature_map_sobel_x = cv2.filter2D(image, -1, sobel_x)
feature_map_sobel_y = cv2.filter2D(image, -1, sobel_y)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('original Image')
plt.axis('off')
plt.subplot(1,4,2)
plt.imshow(feature_map_sobel_x,cmap='gray')
plt.title('soble x Feature Map')
plt.axis('off')
plt.subplot(1,4,3)
plt.imshow(feature_map_sobel_y, cmap='gray')
plt.title("soble y Feature Map")
plt.axis('off')
plt.tight_layout()
plt.show()
```
